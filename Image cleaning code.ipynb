```python
from PIL import Image
import os

data_dir = "plantvillage dataset/color"# or the path of your dataset
bad_files = []

for root, _, files in os.walk(data_dir):
    for file in files:
        path = os.path.join(root, file)
        try:
            with Image.open(path) as img:
                img.verify()
        except Exception:
            bad_files.append(path)

print(f" {len(bad_files)} corrupted images:")
for f in bad_files:
    print(f)
```
